```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10, 50)
y = x**2
dy = 2*x

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].plot(x, y, color='blue')
axes[0].set_xlabel('X  ---------->', color='black')
axes[0].set_ylabel('y=f(x)=x**2', color='black')
axes[0].set_title("plot for f(x)")
axes[0].grid(True, linestyle='--', linewidth=0.5)
axes[0].tick_params(colors='orange')

axes[1].plot(x, dy, color='red')
axes[1].set_xlabel('X  ---------->', color='black')
axes[1].set_ylabel("y=f'(x)=2x", color='black', labelpad=0.1)
axes[1].set_title("plot for f(x)")
axes[1].grid(True, linestyle='--', linewidth=0.5)
axes[1].tick_params(colors='orange')
```
