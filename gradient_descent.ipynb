```python
import numpy as np
import matplotlib.pyplot as plt

# Define function and derivative
def function(w):
    return w**2

def derivative(w):
    return 2*w

# Generate w values
w_values = np.linspace(-10, 10, 100)
f_values = function(w_values)

# Initialize gradient descent
w = 8  # Start far from the minimum
alpha = 0.2  # Learning rate
iterations = 10
w_history = [w]

for i in range(iterations):
    grad = derivative(w)  # Compute gradient
    w = w - alpha * grad  # Gradient descent update
    w_history.append(w)  # Store for visualization

# Plot function
plt.figure(figsize=(8, 5))
plt.plot(w_values, f_values, label="f(w) = w^2", color="blue")
plt.scatter(w_history, [function(w) for w in w_history], color="red", label="Gradient Descent Steps")
plt.xlabel("w")
plt.ylabel("f(w)")
plt.title("Gradient Descent Visualization")
plt.legend()
plt.grid(True, linestyle='--')
plt.tick_params(colors='orange')
plt.show()
```
