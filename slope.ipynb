```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 50)
```

```python
m1 = 5
m2 = 1
m3 = -5
```

```python
c = 4
```

```python
y1 = m1*x + c
y2 = m2*x + c
y3 = m3*x + c
```

```python
plt.plot(x,y1, color='blue')
plt.plot(x, y2, color='red')
plt.plot(x, y3, color='yellow')
plt.xlabel('X  ---------> ', color='black')
plt.ylabel('y = mX+c', color='black')
plt.title('Multiple linear equations', color='black', fontweight='bold')
plt.grid(True, linestyle="--", linewidth='0.5')
plt.xticks(color='orange')
plt.yticks(color='orange')
plt.show()
```
