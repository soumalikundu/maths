```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define function and its derivative
def f(x):
    return x**2
```

```python
def f_derivative(x):
    return 2*x  # Analytical derivative
```

```python
# Generate x values
x_vals = np.linspace(-3, 3, 100)
y_vals = f(x_vals)
derivative_vals = f_derivative(x_vals)
```

```python
# Choose a point where we'll compute discrete slopes
x0 = 1  
h_values = [1, 0.5, 0.1, 0.01]  # Different step sizes
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label="f(x) = x²", color='blue')

# Plot discrete slopes
for h in h_values:
    x1 = x0 + h
    slope = (f(x1) - f(x0)) / (x1 - x0)
    
    # Tangent line using y = mx + c
    tangent_x = np.array([x0, x1])
    tangent_y = f(x0) + slope * (tangent_x - x0)
    
    plt.plot(tangent_x, tangent_y, label=f"Slope for h={h}", linestyle="dashed")

# Plot continuous derivative
plt.plot(x_vals, derivative_vals, label="f'(x) = 2x (Derivative)", color='red', linestyle="dotted")

# Highlight x0
plt.scatter([x0], [f(x0)], color='black', zorder=3)
plt.text(x0, f(x0) + 1, "x₀", fontsize=12, ha='center')

# Labels and legend
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Discrete Slopes Approaching the Derivative")
plt.legend()
plt.grid(True)
plt.show()
```
